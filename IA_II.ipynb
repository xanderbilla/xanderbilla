```python
#Linear search

n = int(input("Enter a number of elements: "))
x = []
for i in range(n):
    x.append(int(input("Enter an element: ")))
print("The entered list i: ", x) 

key = int(input("Enter key element: "))

for i in range(n):
    if key == x[i]:
        print("The key element {} found at index {}".format(key, i))
        break
else:
    print("Element not found...")  
```

```python
# Binary search

def binary(x, key, low = 0, high = len(x)):
    if low > high:
        return -1
    mid = (high + low) // 2
    if key == x[mid]:
        return mid
    elif key > x[mid]:
        return binary(x, key, mid + 1, high)
    else:
        return binary(x, key, low, high -1)

n = int(input("Enter number of element: "))
x = []
for i in range(n):
    x.append(int(input("Enter an element: ")))
print("The created list is: ", x)

key = int(input("Enter the key element: "))
result = binary(x, key, low = 0, high = len(x))
if result  == -1:
    print("Element {} not found".format(key))
else:
    print("Element {} found at index {}".format(key, result))
```

```python
# Bubble sort

x = []
n = int(input("Enter a number: "))
for i in range(n):
    x.append(int(input("Enter an element: ")))
print("The list is: ", x)

for i in range(n-1, 0, -1):
    for j in range(i):
        if x[j] > x[j+1]:
            x[j], x[j+1] = x[j+1], x[j]

print("The sorted list is: ",x)
```
